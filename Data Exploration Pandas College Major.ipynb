Q4. What College major has the highest mid-career salary ? How much do graduates with this major earn ?

```python
print(df["Mid-Career Median Salary"].max())
print(f"Index for the max mid-career salary : {df['Mid-Career Median Salary'].idxmax()}")
print(df['Undergraduate Major'].loc[8])
```

Q5. Which college major has the lowest starting salary and how much do graduates earn after university?

```python
print(df["Starting Median Salary"].min())
print(f"Index for the min Starting Median salary : {df['Starting Median Salary'].idxmin()}")
print(df['Undergraduate Major'].loc[49])
```

Q6. Which college major has the lowest mid-career salary and how much can people expect to earn with this degree? 

```python
print(df["Mid-Career Median Salary"].min())
print(f"Index for the min mid-career salary : {df['Mid-Career Median Salary'].idxmin()}")
print(df['Undergraduate Major'].loc[18])
```

```python
df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
```

```python
df['Mid-Career 90th Percentile Salary'].subtract(df['Mid-Career 10th Percentile Salary'])
```

```python
spread_col = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
df.insert(1, 'Spread', spread_col)
```

```python
df.head()
```

```python
low_risk = df.sort_values('Spread')
low_risk[['Undergraduate Major', 'Spread']].head()
```

Q7. Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. 

```python
highest_potential = df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

Q8. Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.

```python
highest_spread = df.sort_values('Spread')
highest_spread[['Undergraduate Major', 'Spread']].head()
```

```python
df['Salary Difference'] = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
max_difference_major = df.loc[df['Salary Difference'].idxmax()]

print("Undergraduate Major with the largest difference between high and low earners after graduation:")
print(max_difference_major[['Undergraduate Major', 'Salary Difference']])
```
